```python
import csv
import pandas as pd
import hashlib
```

```python
def getUsername(str):
    str = str.lower()
    str = str.split("/")[1]
    tmp = str.split("-", 1)
    repo = tmp[0]
    str = tmp[1]
    if (str[0:8] == "updated-"):
        str = str.split("-", 1)[1]
    return repo, str
    
str = 'wwu-csci-145/lab1-updated-joshMyersdean-lickball'
getUsername(str)
```

## Repos

```python
usernames = []
usernames_hash = []
labs = []

with open('./data/raw/repositories.csv') as csvfile:
    repos = csv.reader(csvfile, delimiter=',')
    next(repos)
    for row in repos:
        tmp = getUsername(row[0])
        lab = tmp[0]
        username = tmp[1]
        
        labs.append(lab)
        
        usernames.append(username)
        
        hash_object = hashlib.md5(username.encode())
        hex_dig = hash_object.hexdigest()
        usernames_hash.append(hex_dig)

data = pd.read_csv('./data/raw/repositories.csv')
data['username'] = usernames
data['username_hash'] = usernames_hash
data['lab'] = labs
data.to_csv('./data/processed/repo_hash.csv', index=False)
data.head()
```

```python
data = pd.read_csv('./data/processed/repo_hash.csv')
data = data.drop('repo_name', 1)
data = data.drop('username', 1)
data.to_csv('./data/processed/repo_hash.csv', index=False)
data.head()
```

## Username

```python
usernames_hash = []
usernames = []

with open('./data/raw/usernames.csv') as csvfile:
    repos = csv.reader(csvfile, delimiter=',')
    next(repos)
    for row in repos:
        username = row[4].lower()
        usernames.append(username)
        hash_object = hashlib.md5(username.encode())
        hex_dig = hash_object.hexdigest()
        usernames_hash.append(hex_dig)


data = pd.read_csv('./data/raw/usernames.csv')
data['username_hash'] = usernames_hash
data.to_csv('./data/processed/username_hash.csv', index=False)
data.head()
```

```python
data = pd.read_csv('./data/processed/username_hash.csv')
data = data.drop('name', 1)
data = data.drop('username', 1)
data = data.rename({'id': 'id_hash'}, axis=1)
data.to_csv('./data/processed/username_hash.csv', index=False)
data.head()
```
